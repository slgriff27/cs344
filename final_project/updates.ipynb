# Updates

Since the project walkthrough, I have gone through the chapter 3.5 in the book Deep Learning with Python (https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/3.5-classifying-movie-reviews.ipynb), recreated that project in my own Jupyter notebook, and got it working. After that, I duplicated that Jupyter notebook and started making changes to it to transform it into the Amazon electronics reviews project. I loaded the json.gz dataset file from http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Electronics_5.json.gz online, and I read it into a pandas object. Then I calculated the percentage helpfulness so that I could get rid of the 50% of reviews that were unhelpful before feeding it into the network. I also tokenized the words in each review so that they were no longer lists of words, but lists of numbers. That is the only way for them to be passed into the system. After all of this, I split the training and testing datasets. I also manually created the training and testing labels from my own interpretation of what kind of an overall rating should be positive, (neutral), or negative.
