# Project Proposal

My original idea is to look into natural language processing more using a dataset of reviews of Amazon electronics. The link to that explains this dataset is here: <http://jmcauley.ucsd.edu/data/amazon/>. It is the 5-core dataset for electronics, meaning each reviewer is only allowed to write 5 reviews, and each electronic device is only allowed 5 reviews. Each entry in this dataset contains the ID of the reviewer, the ID of the product, the name of the reviewer, the helpfulness rating of the review (e.g. 2/3), the text of the review, the overall rating of the product, the summary of the review, and the time of the review in two different formats. What I would plan to do is look at the text of the review and analyze it to figure out if it was a favorable review or not. The way I can check to see if I am correct is by looking at the overall rating of the product. If I predicted that the reviews were mainly positive for a specific item, then a higher rating would verify that while a lower rating would disprove my prediction.

I would be using a recurrent neural network to carry out this sentiment analysis on the Amazon electronics dataset. This kind of a network can handle inputs of different sizes and can remember them from iteration to iteration. The weights will be changed each time, and they will take both new input and previously calculated input to combine for a better response. As the model continues to train, it will learn what makes up a positive review and what makes up a negative review. This kind of memory will help it see patterns more quickly and make more accurate predictions about the reviews.
